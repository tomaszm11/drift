```python
import numpy as np
import pickle
import matplotlib.pyplot as plt


#I'd like to operate on 3-d arrays. 25 voltages were tested on 2 paths, each with 20 separate measurement series


chanelA13=[[[]for _ in range(20)]for _ in range(25)]
chanelA23=[[[]for _ in range(20)]for _ in range(25)]

chanelB23=[[[]for _ in range(20)]for _ in range(25)]
chanelB13=[[[]for _ in range(20)]for _ in range(25)]
time13=[[[]for _ in range(20)]for _ in range(25)]
time23=[[[]for _ in range(20)]for _ in range(25)]

#reading data from text files
for i in range(25):
    for j in range(20):
        if j in range(9):
            time13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,0])
            chanelA13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,1])
            chanelB13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,2])
        elif j in range(9,20):
            time13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,0])
            chanelA13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,1])
            chanelB13[i][j].append(np.loadtxt('./13_{numer}00/13_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,2])

for i in range(25):
    for j in range(20):
        if j in range(9):
            time23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,0])
            chanelA23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,1])
            chanelB23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_0{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,2])
        elif j in range(9,20):
            time23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,0])
            chanelA23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,1])
            chanelB23[i][j].append(np.loadtxt('./23_{numer}00/23_{numer}00_{n}.txt'.format(numer=i+6,n=j+1),skiprows=2)[:][:,2])


#making sure data format is correct

print(time23[1][0][0])
print(time23[15][0])
print('------------------------------------')
print(chanelA23[1][0])
print(chanelA23[24][0])
```

```python
#saving data into files, to use in a different program (where actual analysis will take place)
cha13=open('chanelA13','wb')
pickle.dump(chanelA13,cha13)
cha13.close()
cha23=open('chanelA23','wb')
pickle.dump(chanelA23,cha23)
cha23.close()
chb13=open('chanelB13','wb')
pickle.dump(chanelB13,chb13)
chb13.close()
chb23=open('chanelB23','wb')
pickle.dump(chanelB23,chb23)
chb23.close()
t13=open('time13','wb')
pickle.dump(time13,t13)
t13.close()
t23=open('time23','wb')
pickle.dump(time23,t23)
t23.close()

#finding mid point of each electron spike

midxA13=[[]for _ in range(25)]
midxB13=[[]for _ in range(25)]
midxA23=[[]for _ in range(25)]
midxB23=[[]for _ in range(25)]
midxxA13=[[[]for _ in range(20)]for _ in range(25)]
midxxB13=[[[]for _ in range(20)]for _ in range(25)]
midxxA23=[[[]for _ in range(20)]for _ in range(25)]
midxxB23=[[[]for _ in range(20)]for _ in range(25)]
a=[[[]for _ in range(20)]for _ in range(25)]
b=[[[]for _ in range(20)]for _ in range(25)]
c=[[[]for _ in range(20)]for _ in range(25)]
d=[[[]for _ in range(20)]for _ in range(25)]

for i in range(25):
    for j in range(20):
        midxA13[i].append(max(chanelA13[i][j][0])/2)
        midxB13[i].append(max(chanelB13[i][j][0])/2)
        midxA23[i].append(max(chanelA23[i][j][0])/2)
        midxB23[i].append(max(chanelB23[i][j][0])/2)
        
#saving those for later analysis
mdA13=open('midA13','wb')
pickle.dump(midxA13,mdA13)
mdA13.close()

mdA23=open('midA23','wb')
pickle.dump(midxA23,mdA23)
mdA23.close()

mdB13=open('midB13','wb')
pickle.dump(midxB13,mdB13)
mdB13.close()

mdB23=open('midB23','wb')
pickle.dump(midxB23,mdB23)
mdB23.close()

```

```python
#we will find mid points'indexes
#max points were found to limit our search to one side of the slopes (left)
#midxx are arrays of differences between actual mid value and data points
maxA13=[[[]for _ in range(20)]for i in range(25)]
maxA23=[[[]for _ in range(20)]for i in range(25)]
maxB13=[[[]for _ in range(20)]for i in range(25)]
maxB23=[[[]for _ in range(20)]for i in range(25)]

for i in range(25):
    for j in range(20):
        maxA13[i][j].append(max(chanelA13[i][j][0]))
        maxA23[i][j].append(max(chanelA23[i][j][0]))
        maxB13[i][j].append(max(chanelB13[i][j][0]))
        maxB23[i][j].append(max(chanelB23[i][j][0]))
for i in range(25):
    for j in range(20):
        for k in range(len(chanelA13[i][j][0])):
            if chanelA13[i][j][0][k]!=maxA13[i][j]:
                midxxA13[i][j].append(abs(chanelA13[i][j][0][k]-midxA13[i][j]))
            if chanelA13[i][j][0][k]==maxA13[i][j]:
                break
            
for i in range(25):
    for j in range(20): 
        for k in range(len(chanelA23[i][j][0])):
            if chanelA23[i][j][0][k]!=maxA23[i][j]:
                midxxA23[i][j].append(abs(chanelA23[i][j][0][k]-midxA23[i][j]))
            if chanelA23[i][j][0][k]==maxA23[i][j]:
                break
for i in range(25):
    for j in range(20): 
        for k in range(len(chanelB13[i][j][0])):
            if chanelB13[i][j][0][k]!=maxB13[i][j]:
                midxxB13[i][j].append(abs(chanelB13[i][j][0][k]-midxB13[i][j]))
            if chanelB13[i][j][0][k]==maxB13[i][j]:
                break
for i in range(25):
    for j in range(20): 
        for k in range(len(chanelB23[i][j][0])):
            if chanelB23[i][j][0][k]!=maxB23[i][j]:
                midxxB23[i][j].append(abs(chanelB23[i][j][0][k]-midxB23[i][j]))
            if chanelB23[i][j][0][k]==maxB23[i][j]:
                break
#copying and sorting midxx arrays to find 3 smallest differences
for i in range(25):
    for j in range(20):
        a[i][j]=midxxA13[i][j].copy()
        b[i][j]=midxxB13[i][j].copy()
        c[i][j]=midxxA23[i][j].copy()
        d[i][j]=midxxB23[i][j].copy()
        a[i][j].sort()
        b[i][j].sort()
        c[i][j].sort()
        d[i][j].sort()

```

```python
#poz are mid position arrays. They will contain 3 points of smallest (midxx and data point) differences for minimizing --
#-- later mistakes
pozA13=[[[[],[],[]]for _ in range(20)]for _ in range(25)]
pozB13=[[[[],[],[]]for _ in range(20)]for _ in range(25)]
pozA23=[[[[],[],[]]for _ in range(20)]for _ in range(25)]
pozB23=[[[[],[],[]]for _ in range(20)]for _ in range(25)]

for i in range(25):
     for j in range(20):
        for k in range(len(midxxA13[i][j])):
            if round(a[i][j][0],16)==round(midxxA13[i][j][k],16):
                pozA13[i][j][0].append(k)
            if round(a[i][j][1],16)==round(midxxA13[i][j][k],16):
                pozA13[i][j][1].append(k)
            if round(a[i][j][2],16)==round(midxxA13[i][j][k],16):
                pozA13[i][j][2].append(k)
        for k in range(len(midxxB13[i][j])):
            if round(b[i][j][0],16)==round(midxxB13[i][j][k],16):
                pozB13[i][j][0].append(k)
            if round(b[i][j][1],16)==round(midxxB13[i][j][k],16):
                pozB13[i][j][1].append(k)
            if round(b[i][j][2],16)==round(midxxB13[i][j][k],16):
                pozB13[i][j][2].append(k)

        for k in range(len(midxxA23[i][j])):
            if round(c[i][j][0],16)==round(midxxA23[i][j][k],16):
                pozA23[i][j][0].append(k)
            if round(c[i][j][1],16)==round(midxxA23[i][j][k],16):
                pozA23[i][j][1].append(k)
            if round(c[i][j][2],16)==round(midxxA23[i][j][k],16):
                pozA23[i][j][2].append(k)

        for k in range(len(midxxB23[i][j])):
            if round(d[i][j][0],16)==round(midxxB23[i][j][k],16):
                pozB23[i][j][0].append(k)
            if round(d[i][j][1],16)==round(midxxB23[i][j][k],16):
                pozB23[i][j][1].append(k)
            if round(d[i][j][2],16)==round(midxxB23[i][j][k],16):
                pozB23[i][j][2].append(k)
#saving positions
pA13=open('pozA13','wb')
pickle.dump(pozA13,pA13)
pA13.close()
pB13=open('pozB13','wb')
pickle.dump(pozB13,pB13)
pB13.close()
pA23=open('pozA23','wb')
pickle.dump(pozA23,pA23)
pA23.close()
pB23=open('pozB23','wb')
pickle.dump(pozB23,pB23)
pB23.close()
```
